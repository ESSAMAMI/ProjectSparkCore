```python
import chart_studio.plotly as py
import plotly.figure_factory as ff
import pandas as pd

topClients = pd.read_csv("DataGenerated/topClients.csv", sep=";", encoding="ISO-8859-1")

table = ff.create_table(topClients[:100])
py.iplot(table, filename='jupyter-table1')
```

```python
import plotly.express as px

fig = px.bar(topClients[:50], y='Profits', x='Name', text='Profits')
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.update_layout(uniformtext_minsize=10, uniformtext_mode='hide', title='<b>TOP 50 CLIENTS</b>')
fig.show()
```

```python
bestItems = pd.read_csv('DataGenerated/bestItems.csv', sep=";", encoding="ISO-8859-1")
table = ff.create_table(bestItems[:100])
py.iplot(table, filename='jupyter-table1')
```

```python
fig = px.pie(bestItems, values='Total', names='Category', title='<b>TOP SELLING ITEMS / CATEGORY</b>')
fig.show()
```

```python
bestSales = pd.read_csv('DataGenerated/bestSales.csv', sep=";", encoding="ISO-8859-1")
table = ff.create_table(bestSales[:100])
py.iplot(table, filename='jupyter-table1')
```

```python
import plotly.express as px

fig = px.bar_polar(bestSales[:50], r="Amount", theta="Item_Name",
                   color="Total_Items", template="plotly_dark",
                   color_discrete_sequence= px.colors.sequential.Plasma_r, title='<b>TOP EARNERS ITEMS</b>')
fig.show()
```
